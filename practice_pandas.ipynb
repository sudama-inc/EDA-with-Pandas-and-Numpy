# loc & iloc
*   iloc: use indexing, last not include & loc: use name, last is included


```python
df
```

```python
print(df['col1'])
# To get the particular row it will not work --> so will use the loc & iloc
# print(df['in6'])
# print(df['in6']['col1'])
# print(df[0][1])
```

```python
# To get the particuar row with all columns : loc
print(df.loc['in6'])

# Get the particular row : iloc
print(df.iloc[0])
```

```python
# To get the row from in3 to in6 (range of rows) with all columns
# loc
display(df.loc['in3':'in6'])
display(df.loc['in3':])
# display(df.loc[['in3':]])   # --> WRONG

# iloc
print('Using iloc \n')
display(df.iloc[2:6])
display(df.iloc[2:])
```

```python
# Get the multiple rows with all columns
display(df.loc[['in2', 'in4', 'in6']])
print('Using iloc \n')
display(df.iloc[[1, 3, 5]])

# It will through an error
# df.loc['in2', 'in4', 'in6']
# df.iloc[1, 3, 5]
```

```python
# Get single column with all rows
display(df.loc[:, 'col2'])
print('Using iloc \n')
display(df.iloc[:, 1])
```

```python
# Get the from col2 to col5 (range of columns) with all the rows
display(df.loc[:, 'col2':'col5'])
print('Using iloc \n')
display(df.iloc[:, 1:4])
```

```python
# Get the specified multiple columnd
# df.loc[:, 'col1', 'col3', 'col5'] # it will through the error
display(df.loc[:, ['col1', 'col3', 'col5']])
print('Using iloc \n')
display(df.iloc[:, [0, 2, 4]])
```

```python
# Get the specified rows and specified columns
print(df.loc['in1', 'col2'])
print(df.loc['in1']['col2'])
print('Using iloc \n')
print(df.iloc[0, 1])
print(df.iloc[0][1])
```

```python
# Get the specified multiple rows and specified multiple columns
print(df.loc[['in1', 'in3', 'in6'],['col2', 'col3', 'col6']])
print(df.loc[['in1', 'in3', 'in6']][['col2', 'col3', 'col6']])

# It will through an error --> Its not a syntatically correct
# print(df.loc['in1', 'in3', 'in6']['col2', 'col3', 'col6'])

print('Using iloc \n')
print(df.iloc[[0,2,5],[1,2,5]])
print(df.iloc[[0,2,5],[1,2,5]])

# It will not work
# print(df.iloc[[0,2,5]][[1,2,5]])
```

```python
# Get the range of rows and range of columns
df.loc['in1':'in4', 'col2':'col5']
# df.loc[['in1':'ind4']][['col1':'col4']] #----- WRONG
# df.loc[['in1':'ind4'],['col1':'col4']] #----- WRONG

print('Using iloc \n')
df.iloc[0:3, 1:4]
```

```python
student_data={'Name':['Nilima','Namitha', 'Mahesh','Girish','Sonali','Sunil','Deepak','Arush','Mahima', 'Bhushan'],'Score':[76,68,90,92,63,86,88,77,65,89]}
```

```python
st = pd.DataFrame(student_data)
st
```

```python
# grade: score>80 :A grade || 60<score<80: B grade || score<60: C grade
st['grade'] = st.apply(lambda x : 'A' if (x['Score']>=80) else 'B' if ((x['Score']<80) and (x['Score']>=60)) else 'C', axis=1)
st
```

```python
sales1={'item':['item1','item2','item3','item4'], 'MRP':[2000,2503,4702,3564], 'disc':[0.1,0.2,0.25,0.15]}
sales = pd.DataFrame(sales1)
sales
```

```python
# Discounted Price : Create a column as disc formula for : [mrp-(mrp*disc)]

sales['dis_price'] = sales.apply(lambda x : x.MRP - x.MRP*x.disc, axis=1)
sales
```

```python
# Create another column named "10_perc_disc_price" for all the items
sales['10_perc_disc_price'] = sales.apply(lambda x: x['MRP'] - x['MRP']*.10, axis=1)
sales
```

## Merging, Joining, and Concatenating

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
df4 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'],
                        'F': ['B4', 'B5', 'B6', 'B7'],
                        'G': ['C4', 'C5', 'C6', 'C7'],
                        'H': ['D4', 'D5', 'D6', 'D7']},
                         index=[0, 1, 2, 3])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
df5 = pd.DataFrame({'I': ['A8', 'A9', 'A10', 'A11'],
                        'J': ['B8', 'B9', 'B10', 'B11'],
                        'K': ['C8', 'C9', 'C10', 'C11'],
                        'L': ['D8', 'D9', 'D10', 'D11']},
                        index=[0,1,2,3])
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis=1)
```

### Merging

```python
lft = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

rht = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K4'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(lft, rht, how='inner', on='key')
```

```python
pd.merge(lft, rht, how='left', on='key')
```

```python
pd.merge(lft, rht, how='right', on='key')
```

## Joining
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```python
lft = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

rht = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3'],
                    'E': ['E0', 'E2', 'E3']},
                      index=['K0', 'K2', 'K3'])
```

```python
lft.join(rht)
```

```python
lft.join(rht,on='A', how='outer', )
```
