***Let's Implement loc***

```python
df1.loc['ind1']     # Print index1 with All column
```

```python
df1.loc[['ind1', 'ind4']]   # Print index1 & index4 with All column
```

```python
df1.loc['ind1' :'ind4']      # Print from index1 till index4 with All column
```

```python
df1.loc['ind1'] ['col1']       # Print the value for index1 & Column1
```

```python
df1.loc['ind1', 'col3']     # Print the value for index1 & Column3
```

```python
df1.loc['ind1':'ind4']['col3']      # Print the value for index1 till  index4 with Column3
```

```python
df1.loc[['ind1', 'ind3']]['col6']     # Print the value for index1 and index3 with Column6
```

```python
df1.loc[['ind1','ind3']][['col3', 'col4']]      # Print the value for index1 and index3 with column3 and column4
```

```python
df1.loc['ind3':'ind6', 'co2':'col5']      # Print the value for index3 till index6 with column2 till column5
```

```python
df1.loc[:, 'col3':'col5']      # Print the value of all index with column3 and column5
```

```python
student_data={'Name':['Nilima','Namitha', 'Mahesh','Girish','Sonali','Sunil','Deepak','Arush','Mahima', 'Bhushan'],
              'Score':[76,68,90,92,63,86,88,77,65,89]}
student_data
```

```python
# Convert this to the DataFrame

st=pd.DataFrame(student_data)
st
```

```python
# grade: score>80 :A grade
# score is below 80: B grade
```

```python
st['Grade'] = st.apply(lambda x:'A' if x.Score > 80 else 'B', axis=1)
st
```

```python
### Create another column grade1 where grade A: >80; B>70 and <80; grade C is <70

st['grade1']=st.apply(lambda x: 'A' if x.Score > 80 else ('B' if x.Score > 70 and  x.Score < 80 else 'C'), axis=1)
st
```

```python
sales1={'item':['item1','item2','item3','item4'], 'MRP':[2000,2503,4702,3564], 'disc':[0.1,0.2,0.25,0.15]}

# Convert this Sales1 into DataFrame
sales = pd.DataFrame(sales1)
sales
```

```python
sales['Discounted_price']=sales.apply(lambda x: x.MRP - x.MRP*x.disc, axis=1)
sales
```

```python
### Create another column named "10_perc_disc_price" for all the items

sales['10_perc_disc_price']=sales.apply(lambda x: x.MRP - x.MRP*x.disc, axis=1)
```
