#### Read the file with pandas parquet

#### Reading and writing Pickle

```python
# Reading and writing Pickle
pickle_file = 'creditcard.pickle'
df.to_pickle(file_path + pickle_file)
df_pickle = pd.read_pickle(file_path + pickle_file)
df_pickle
```

```python
# Reading and writing Feather
feather_file = 'creditcard.feather'
df.to_feather(file_path + feather_file)
df_feather = pd.read_feather(file_path + feather_file)
df_feather
```

```python
chunk_size = 100  # Number of rows per chunk
total_entries = 500  # Total number of entries to retrieve
total_iterations = total_entries // chunk_size  # Calculate the total number of iterations based on chunk size

chunks = []  # List to store the chunks

# Read the dataset in the specified number of iterations
for i in range(total_iterations):
    # Calculate the start and end rows for the current iteration
    start_row = i * chunk_size
    end_row = start_row + chunk_size

    # Read the chunk for the current iteration
    chunk = pd.read_csv('/content/drive/My Drive/Datasets/creditcard.csv', skiprows=range(1, start_row), nrows=chunk_size)

    chunks.append(chunk)  # Append the chunk to the list

# Concatenate all the chunks into a single DataFrame
chunk_df = pd.concat(chunks, ignore_index=True)

# Process the concatenated DataFrame as needed
display(chunk_df)  # Example: printing the chunk for demonstration purposes
```
