```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder, MinMaxScaler, StandardScaler
```

```python
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df.head()
```

###### **Apply MimMaxScaler**

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df)
```

```python
scaled_df = pd.DataFrame(scaled, columns=df.columns)
scaled_df.head(3)
```

###### **Apply StandardScaler**

```python
ss = StandardScaler()
ss = ss.fit_transform(df)
```

```python
ss_df = pd.DataFrame(ss, columns=df.columns)
ss_df.head(3)
```

```python
encoder_df = pd.DataFrame({'eid':[1,2,3,4,5,6,7,8,9],'gender':['m','f','f','m','m','f','m','m','f'],'remark':['a','b','c','a','b','c','a','b','c']})
encoder_df.head(2)
```

###### **Apply get_dummies**

```python
ohe_df = pd.get_dummies(encoder_df, columns=['gender', 'remark'], drop_first=True)
```

###### **Apply OneHotEncoder**

```python
scale = OneHotEncoder(drop='first')
scaled_df = pd.DataFrame(scale.fit_transform(encoder_df[['gender','remark']]).toarray())
scaled_df
```

```python
combined_df = encoder_df[['eid']].join(scaled_df)
combined_df
```

###### **Apply OrdinalEncoder**

```python
oe = OrdinalEncoder()
oe.fit(encoder_df[['gender','remark']])
oe.transform(encoder_df[['gender','remark']])
```

###### **Apply LabelEncoder**

```python
edf = encoder_df.copy()
le = LabelEncoder()
edf['gender'] = le.fit_transform(edf['gender'])
edf['remark'] = le.fit_transform(edf['remark'])
edf
```
